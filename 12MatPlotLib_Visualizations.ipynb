```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df
```

```python
df['species'] = iris.target
```

```python
df
```

```python
df['species_name'] = df['species'].apply(lambda x: iris.target_names[x])
```

```python
species_counts = df['species_name'].sample(n=50).value_counts()
species_counts
```

```python
plt.figure(figsize=(4, 3))
plt.bar(species_counts.index, species_counts.values, color=['tomato', 'green', 'skyblue'])
plt.title('Count of Each Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
# Pie Chart – Proportion of each species
df['species'].sample(n=50).value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Species Distribution")
plt.ylabel("")  # Remove y-label
plt.show()
```

```python
#Histogram – Sepal length
df['sepal length (cm)'].plot(kind='hist', bins=10, color=['red'])
plt.title("Sepal Length Histogram")
plt.xlabel("Sepal Length (cm)")
plt.show()
```

```python
#Scatter Plot – Sepal length vs Sepal width
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], color='green')
plt.title("Sepal Length vs Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
# Box Plot – Petal length
df.boxplot(column='petal length (cm)', by='species_name')
plt.title("Box Plot of Petal Length by Species")
plt.suptitle("")  # Remove extra title
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
#Line Chart – Average of each feature per species

df.groupby('species_name').mean().plot(kind='line', marker='*')
plt.title("Average Feature Values per Species")
plt.xlabel("Species")
plt.ylabel("Value (cm)")
plt.show()
```
