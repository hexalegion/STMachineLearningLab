```python
import seaborn as sns
```

```python
titanic  = sns.load_dataset('titanic')
```

```python
titanic
```

```python
import pandas as pd

survived_frequency = pd.crosstab(titanic['pclass'],titanic['survived'])
```

```python
survived_frequency
```

```python
#Stacked Bar Chart
import matplotlib.pyplot as plt

survived_frequency.plot(kind='bar',stacked=True,color=['Blue','yellow'])
plt.title("Class-wise Survival")
plt.xlabel("Classes")
plt.ylabel("Total number")
```

```python
#Interpretation :
#In class 1, a little more than 50% of the persons survived
```

```python
survived_data=titanic[titanic['survived'] == 1]

sns.kdeplot(survived_data['age'])
```

```python
for label, colour in zip([0,1],['red','green','blue']):
    sns.kdeplot(titanic[titanic['survived']==label]['age'])
```

```python
#Violin Plot

sns.violinplot(x='class',y='age',data=titanic)
```
